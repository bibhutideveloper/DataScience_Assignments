```python
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
data=pd.read_csv("stock_data.csv")
```

```python
data
```

```python
type(data)
```

```python
del data["Date"]
```

```python
del data["Name"]
```

```python
data
```

```python
plt.scatter(data.Open,data.Close)
plt.show()
```

```python
a=KMeans(n_clusters=10)
```

```python
a.fit(data)
```

```python
plt.scatter(data.Open,data.Close)
plt.plot()
```

```python
plt.scatter(data.Open,data.Close,c = a.labels_)
plt.plot()
```

```python
data= [[55,78,56,6,56],[8,32,67,6,32],[65,76,45,4,32],[43,12,4,65,65],[6,22,6,56,43]]
```

```python
points = ["a","b","c","d","e"]
```

```python
df = pd.DataFrame(data,columns=["a","b","c","d","e"], index=points)
df
```

```python
linkage_data = linkage(df, method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.show()
```

```python
linkage_data = linkage(df, method='complete', metric='euclidean')
dendrogram(linkage_data)

plt.show()
```

```python
linkage_data = linkage(df, method='single', metric='euclidean')
dendrogram(linkage_data)

plt.show()
```

```python
linkage_data = linkage(df, method='average', metric='euclidean')
dendrogram(linkage_data)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv("stock_data.csv")
```

```python
del df1["Name"]
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(df1)
data_scaled = pd.DataFrame(data_scaled, columns=df1.columns)
data_scaled.head()
```

```python
df1
```

```python
del df1["Date"]
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("HCA Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
```

```python
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.axhline(y=6, color='r', linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')  
cluster.fit_predict(data_scaled)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(data_scaled['Open'], data_scaled['Close'], c=cluster.labels_)
```
